```python
import pandas as pd
```

```python
aug_19 = pd.read_csv("Data/201908-citibike-tripdata.csv")
sept_19 = pd.read_csv("Data/201909-citibike-tripdata.csv")
oct_19 = pd.read_csv("Data/201910-citibike-tripdata.csv")

aug_20 = pd.read_csv("Data/202008-citibike-tripdata.csv")
sept_20 = pd.read_csv("Data/202009-citibike-tripdata.csv")
oct_20 = pd.read_csv("Data/202010-citibike-tripdata.csv")


aug_19.head()
```

```python
# combine all csv files into 1 db

combined_df = pd.concat([aug_19, sept_19, oct_19, aug_20, sept_20, oct_20])

combined_df.head()
```

```python
# drop columns
combined_df.drop(columns=['bikeid', 'stoptime', 'start station id', 'end station id'], inplace=True)



```

```python
# change gender to male / female
combined_df["gender"].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}, inplace=True)

combined_df.head()
```

```python
#break out year

combined_df['Trip Year'] = combined_df.starttime.str.split("-", expand=True)[0]
combined_df.head()
```

```python
#break out month
combined_df['Trip Month'] = combined_df.starttime.str.split("-", expand=True)[1]
combined_df.head()

combined_df["Trip Month"].replace({"08": "August", "09": "September", "10": "October"}, inplace=True) 
combined_df
```

```python
#save new csv

combined_df.to_csv(r'Data/clean_df.csv')
```
