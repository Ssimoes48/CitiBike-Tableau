```python
# clean_df2.drop(columns=['Unnamed: 0'], inplace=True)

clean_df3 = clean_df2[["Trip Year", "Trip Month", "Start Station", "Start Latitude", "Start Longitude", "End Station", "End Latitude", "End Longitude", "Trip Duration", "User Type", "Rider Gender", "Rider Age", "Birth Year", "Start Time"]]
clean_df3.head()
```
