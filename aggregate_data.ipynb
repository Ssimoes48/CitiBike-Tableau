```python
import pandas as pd
```

```python
clean_df3 = pd.read_csv("Data/clean_df3.csv")
clean_df3.head()
```

```python
clean_df3.drop(columns=['Unnamed: 0'], inplace=True)
# clean_df3["Rider Gender"].replace({'na': 'NaN'}, inplace=True)
clean_df3.head(100)
```

```python
months = clean_df3["Trip Month"].unique()

months
```

```python
month_df = clean_df3.groupby(['Trip Year','Trip Month']).count()

# month_final_df = month_df[["Trip Year", "Trip Month", "Start Station"]]

month_df.drop(columns=['Start Latitude', 'Start Longitude', 'End Station', 'End Latitude', 'End Longitude', 'Trip Duration', 'User Type', 'Rider Gender', 'Rider Age', 'Birth Year', 'Start Time' ], inplace=True)


month_df.rename(columns={'Start Station' : 'Total Trips'})

month_df
```

```python
month_df.to_csv(r'Data/month_df.csv')
```

```python
location_df = clean_df3.groupby(["Start Station", "Start Latitude", "Start Longitude"]).count()

location_df
```

```python
location_df.to_csv(r'Data/start_location.csv')
```

```python
end_location_df = clean_df3.groupby(["End Station", "End Latitude", "End Longitude"]).count()

end_location_df 
```

```python
end_location_df.to_csv(r'Data/end_location.csv')
```

```python
station_df = clean_df3.groupby(['Start Station']).agg({"Trip Duration": "mean"}).round(2).rename(columns={'Trip Duration' : 'Average Trip Duration'}) 

station_df
```

```python
station_df.to_csv(r'Data/station_mean_trip.csv')
```
